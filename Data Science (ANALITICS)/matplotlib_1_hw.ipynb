#### Описание данных

Для выполнения домашнего задания необходимо использовать датасет `vgsales.csv`, загрузка которого будет приведена ниже. Этот датасет содержит информацию о суммарных продажах видеоигр для игровых консолей. Описание признаков:

* `EU_Sales` – объем продаж игры в Европе, млн. копий
* `JP_Sales` – объем продаж игры в Японии, млн. копий
* `Other_Sales` – объем продаж игры в остальных странах, млн. копий
* `Global_Sales` – объем продаж игры по всему миру, млн. копий

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Загрузка данных

```python
df = pd.read_csv('vgsales.csv')
df.head()
```

### Задача 1

Постройте гистограмму распределения объема продаж всех игр в Японии.

```python
serie = df.groupby('Year')['JP_Sales'].count()
plt.hist(serie, bins=10)
plt.title("Гистограмма распределения объема продаж всех игр в Японии")
```

### Задача 2

Постройте столбчатую диаграмму (barchart), показывающую сравнение общемирового объема продаж игры "Grand Theft Auto V" на разных платформах.

```python
df1 = df[df['Name'] == "Grand Theft Auto V"]
plt.bar(df1['Platform'], df1['Global_Sales'])
plt.title("Общемировой объем продаж игры 'Grand Theft Auto V' на разных платформах")
plt.ylabel('Объем продаж, млн. копий')
plt.xlabel('Платформа')
```

### Задача 3

Постройте линейный график динамики суммарных общемировых продаж всех игр по годам.

```python

series = df.groupby('Year')['Global_Sales'].sum()
plt.title("Динамики суммарных общемировых продаж всех игр по годам")
plt.xlabel('Год')
plt.ylabel('Объем продаж, млн. копий')
plt.plot(series)
```

### Задача 4

Постройте точечную диаграмму, показывающую существование зависимости продаж во всем мире от продаж в Северной Америке.

```python
plt.scatter(df['NA_Sales'], df['Global_Sales'])
plt.title("Зависимость продаж во всем мире от продаж в Северной Америке")
plt.ylabel('Продажи во всем мире, млн')
plt.xlabel('Продажи в Северной Америке, млн')
```

### Задача 5

Постройте круговую диаграмму процентного соотношения продаж игры "Super Mario Bros." на разных платформах во всем мире. Подпишите доли графика с точностью 1 знак после запятой.

```python
serie = df[df['Name']=='Super Mario Bros.'].groupby('Platform')['Global_Sales'].sum()
plt.pie(serie, labels=serie.index, autopct='%1.1f%%')
plt.title('Распределение продаж игры "Super Mario Bros." по платформам')
plt.show()
```

### Задача 6

Определите 5 издателей, имеющих наибольшие суммарные продажи во всём мире в 2013 году, и проиллюстрируйте величину их продаж на столбчатой диаграмме.

```python
df2 = df[df['Year'] == 2013].sort_values(['Global_Sales', 'Publisher'], ascending=False).head(7)


plt.bar(df2['Publisher'], df2['Global_Sales'])
plt.title("5 издателей, имеющих наибольшие суммарные продажи во всём мире в 2013 году")
plt.xlabel('Издатели')
plt.ylabel('Объем продаж, млн. копий')
plt.xticks(rotation=90)
```

### Задача 7

Постройте гистограмму распределения величины общемировых продаж игр, выпущенных не издателем Nintendo, в период с 2000 по 2015 год включительно. Гистограмма для большей точности должна содержать 20 столбцов, отсечки на оси X должны соответствовать границам столбцов.

```python
plt.figure(figsize=(12,4))
serie = df[(df['Publisher'] != 'Nintendo') & (df['Year'].isin(range(2000, 2016)))]['Global_Sales']
bins_cnt = 20
x_min = serie.min()
x_max = serie.max()
bin_width = (x_max - x_min) / bins_cnt
x_range = np.arange(x_min, x_max+bin_width, bin_width)
_, _, patches = plt.hist(serie, bins=bins_cnt)
plt.xticks(x_range, rotation=45)
plt.title('Распределение общемировых продаж игр, выпущенных не издателем Nintendo,\nв период с 2000 по 2015 год включительно')
plt.xlabel('Объем продаж, млн. копий')
plt.ylabel('Кол-во игр')
plt.show()
```

### Задача 8

Постройте линейный график динамики суммарных мировых продаж игр жанра "Action" по годам.

```python
series = df[df['Genre'] == 'Action'].groupby('Year')['Global_Sales'].sum()

plt.title("Динамики суммарных общемировых продаж игр 'Action' по годам")
plt.xlabel('Год')
plt.ylabel('Объем продаж, млн. копий')
plt.plot(series)
```

### Задача 9

С помощью столбчатой диаграммы проиллюстрируйте объем продаж всех игр издателя "Microsoft Game Studios" в Северной Америке за все время в зависимости от жанра. Столбцы расположите по убыванию.

```python
df2 = df[df['Publisher'] == 'Microsoft Game Studios'].sort_values(['NA_Sales', 'Year'], ascending=False)
plt.bar(df2['Genre'], df2['NA_Sales'])
plt.title("Oбъем продаж всех игр издателя 'Microsoft Game Studios' в Северной Америке")
plt.xlabel('Жанр')
plt.ylabel('Объем продаж, млн. копий')
plt.xticks(rotation=90)
```
