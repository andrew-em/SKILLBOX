```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Загрузка данных

```python
df = pd.read_csv('vgsales.csv')
df.head()
```

### Задача 1

Постройте линейный график динамики объема продаж всех игр жанра "Sports" в Японии. Задайте графику размер 12 на 6

```python
graf = df[df['Genre'] == 'Sports'].groupby('Year')['JP_Sales'].sum()

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(graf)
ax.set_title('Динамика объема продаж всех игр жанра "Sports" в Японии')
ax.set_xlabel('Год')
ax.set_ylabel('Объем продаж, млн. копий')

plt.show()
```

### Задача 2

В одних координатных осях постройте линейные графики динамики продаж студии "Activision" в Северной Америке, Европе, Японии и всем мире

```python
graf_na = df[df['Publisher'] == 'Activision'].groupby('Year')['NA_Sales'].sum()
graf_jap = df[df['Publisher'] == 'Activision'].groupby('Year')['JP_Sales'].sum()
graf_eu = df[df['Publisher'] == 'Activision'].groupby('Year')['EU_Sales'].sum()
graf_world = df[df['Publisher'] == 'Activision'].groupby('Year')['Global_Sales'].sum()

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(graf_eu, label='EUROPE')
ax.plot(graf_na, label='NORTH AMERICA')
ax.plot(graf_jap, label='JAPAN')
ax.plot(graf_world, label='WORLD')

ax.set_title('Динамика продаж студии "Activision" в Северной Америке, Европе, Японии и всем мире')
ax.set_xlabel('Год')
ax.set_ylabel('Объем продаж, млн. копий')

ax.legend(loc='upper left')
plt.show()
```

### Задача 3

В цикле (т.е. на разных координатных осях) постройте линейные графики динамики продаж студии "Activision" в Северной Америке, Европе, Японии и всем мире

```python
for sales in ['NA_Sales', 'JP_Sales',  'EU_Sales', 'Global_Sales']:
    graf_na = df[df['Publisher'] == 'Activision'].groupby('Year')[sales].sum()

    fig, ax = plt.subplots(figsize=(12, 6))
     
    ax.plot(graf_na, label=sales)

    ax.set_title('Динамика продаж студии "Activision" в ' + sales)
    ax.set_xlabel('Год')
    ax.set_ylabel('Объем продаж, млн. копий')

    ax.legend(loc='upper left')
    plt.show()
```

### Задача 4

Постройте такие же линейные графики динамики продаж студии "Activision" в Северной Америке, Европе, Японии и всем мире на одном полотне в сетке 2 на 2 графика. Итоговое полотно сделайте размером 12 на 12. Все графики должны разделять оси X и Y

```python
data = df[df['Publisher'] == 'Activision'].groupby('Year').sum()

fig, axs = plt.subplots(nrows = 2, ncols = 2, sharex = True, sharey = True, figsize = (12, 12))

regions = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Global_Sales']

for idx, region in enumerate(regions):
    ax = axs[idx // 2][idx % 2]
    ax.plot(data.index, data[region])
    ax.set_ylabel('{}, млн. копий'.format(region))
    ax.set_yticks(range(0, 81, 20))
    if idx // 2 == 1:
        ax.set_xlabel('Год')


fig.suptitle('Динамика продаж студии Activision', y = 0.9)

plt.show()
```

### Задача 5

В одной координатной сетке постройте 2 гистограммы распределения мировых продаж игр издателей "Microsoft Game Studios" и "Take-Two Interactive" после 2010 года (включая 2010 год). Графики сделайте полупрозрачными, отсечки должны быть общими и соответствовать ширине столбцов обоих графиков

```python
graf_1 = df[(df['Publisher'] != 'Microsoft Game Studios') & (df['Year'].isin(range(2010, 2016)))]['Global_Sales']
graf_2 = df[(df['Publisher'] != 'Take-Two Interactive') & (df['Year'].isin(range(2010, 2016)))]['Global_Sales']

fig, ax = plt.subplots(figsize = (15,6))
_, bins_1, _ = ax.hist(graf_1, label = 'Microsoft Game Studios', alpha=0.5, bins=10)
ax.hist(graf_2, label = 'Take-Two Interactive', alpha=0.5, bins=bins_1)

ax.set_xticks(list(bins_1))
ax.tick_params(axis = 'x', rotation = 45)

ax.set_title(' Распределения мировых продаж игр издателей "Microsoft Game Studios" и "Take-Two Interactive" после 2010')
ax.set_xlabel('Объем продаж, млн. копий')
ax.set_ylabel('Кол-во игр')
ax.legend()

plt.show()
```
